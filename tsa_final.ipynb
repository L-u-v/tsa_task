```python
#defining columns with different names & skipping row 1
col=['DATE','TIME','CO_GT','PT08_S1_CO','NMHC_GT','C6H6_GT','PT08_S2_NMHC',
     'NOX_GT','PT08_S3_NOX','NO2_GT','PT08_S4_NO2','PT08_S5_O3','T','RH','AH']
df1=pd.read_csv('Test.csv',header=None,skiprows=1,names=col)
df1.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Drop rows with missing target values
df1.dropna(subset=['CO_GT', 'PT08_S1_CO', 'NMHC_GT', 'C6H6_GT', 'PT08_S2_NMHC', 'NOX_GT',
       'PT08_S3_NOX', 'NO2_GT', 'PT08_S4_NO2', 'PT08_S5_O3', 'T', 'RH', 'AH'], inplace=True)
```

```python
# Split the data into features and target variables
X = df1.drop(columns=['CO_GT', 'NMHC_GT', 'C6H6_GT', 'NOX_GT', 'NO2_GT', 'HOUR'])
y = df1[['CO_GT', 'PT08_S1_CO', 'NMHC_GT', 'C6H6_GT', 'PT08_S2_NMHC', 'NOX_GT',
       'PT08_S3_NOX', 'NO2_GT', 'PT08_S4_NO2', 'PT08_S5_O3', 'T', 'RH', 'AH',
       'HOUR']]
```

```python
df1.columns
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize and train the Random Forest Regressor model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)
```

```python
# Evaluate RMSE for each pollutant
rmse = {}
for i, col in enumerate(y.columns):
    rmse[col] = sqrt(mean_squared_error(y_test.iloc[:, i], predictions[:, i]))
# Check if RMSE meets the criteria
criteria_met = all(rmse[col] <= threshold for col, threshold in [
    ('CO_GT', 1.3),
    ('PT08_S1_CO', 175),
    ('NMHC_GT', 4),
    ('C6H6_GT', 5),
    ('PT08_S2_NMHC', 250),
    ('NOX_GT', 175),
    ('PT08_S3_NOX', 210.73),
    ('NO2_GT', 50.17),
    ('PT08_S4_NO2', 300.7),
    ('PT08_S5_O3', 400.25),
    ('T', 6.7),
    ('RH', 15.5),
    ('AH', 0.3)
])

if criteria_met:
    print("RMSE criteria met for all pollutants.")
else:
    print("RMSE criteria not met for all pollutants.")

# Save the predictions to a submission file (submission.xlsx)
submission = pd.DataFrame(predictions, columns=y.columns)
submission.to_excel('submission.xlsx', index=False)
```
